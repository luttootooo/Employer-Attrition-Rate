```python
def isi(x):
    if x==19 or x==20 or x==21 or x==22 or x==23 or x==24 or x==25 :
        return (x-19)
    elif x==26 or x==27 or x==28 or x==29 or x==30:
        return random.randint(x-24,x-22)
    elif 31<=x<=39: 
        return random.randint(x-29,x-22)
    elif x==40: 
        return random.randint(11,18)
    elif 41<=x<=65:
        return random.randint(x-39,x-22)
    else:
        return random.randint(0,43)
```

```python
def gender(x):
    if x=='F':
        return 0
    elif x=='M':
        return 1

df['Gender']=df['Gender'].apply(gender)
```

```python
def status(x):
    if x=='Single':
        return 0
    elif x=='Married':
        return 1

df['Relationship_Status']=df['Relationship_Status'].apply(status)
```

```python
def unit(x):
    if x=='IT':
        return 0
    elif x=='Logistics':
        return 1
    elif x=='Sales':
        return 2
    elif x=='Operarions':
        return 3
    elif x=='R&D':
        return 4
    elif x=='Purchasing':
        return 4
    elif x=='Accounting and Finance':
        return 5
    elif x=='Human Resource Management':
        return 6
    elif x=='Marketing':
        return 7
    elif x=='Production':
        return 8
    elif x=='Quality':
        return 9
    elif x=='Security':
        return 10

df['Unit']=df['Unit'].apply(unit)
```

```python
def tipe(x):
    if x=='type0':
        return 0
    elif x=='type1':
        return 1
    elif x=='type2':
        return 2
    elif x=='type3':
        return 3
    elif x=='type4':
        return 4

df['Compensation_and_Benefits']=df['Compensation_and_Benefits'].apply(tipe)
```
